## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
ax= sns.boxplot(df["ConvertedComp"], orient="v")
```

```python
ax = sns.boxplot(x=df['Age'], orient="v")
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
IQR = df['ConvertedComp'].quantile(0.75) - df['ConvertedComp'].quantile(0.25)
print("The inter Quartile Range is:", IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
Lbound = df['ConvertedComp'].quantile(0.25) - 1.5*IQR
Lbound = max(Lbound, df['ConvertedComp'].min())
print("The Lower Bound is:", Lbound)

Ubound = df['ConvertedComp'].quantile(0.75) + 1.5*IQR
print("The Upper Bound is:", Ubound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
print("The number of outliers in ConvertedComp is:",len(df[df['ConvertedComp'] > Ubound]))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df2 = df[df['ConvertedComp'] <= Ubound]
df2['ConvertedComp'].median()
```

```python
df2['ConvertedComp'].mean()
```
